### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
df.describe(include='all')
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing_values=df.isnull().sum()
missing_values
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
sns.heatmap(df.isnull())
plt.title("Heatmap of missing values")

# Sort columns by the number of missing values
sorted_cols = missing_values.sort_values(ascending=False).index

plt.figure(figsize=(15, 10))
sns.heatmap(df[sorted_cols].isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Heatmap of Missing Values (Sorted by Missing Count)', fontsize=16)
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_values_employment=df['Employment'].isnull().sum()
print("Number of missing values in employment:",missing_values_employment)
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
print("Most Frequently occuring value in Employment is:",df['Employment'].value_counts().idxmax())
print("Most Frequently occuring value in columns")
for col in df.columns:
    print(col,":",df[col].value_counts().idxmax())
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
df['Employment'].fillna(df['Employment'].value_counts().idxmax(),inplace=True)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
missing_RemoteWork=df['RemoteWork'].isnull()
print(df['RemoteWork'].value_counts())
frequent_RemoteWork=df['RemoteWork'].value_counts().idxmax()
df['RemoteWork'].fillna(frequent_RemoteWork,inplace=True)
print(df['RemoteWork'].value_counts())
```
