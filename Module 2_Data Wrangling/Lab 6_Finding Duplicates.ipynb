### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here
#i=0
#Column=[]
#Duplicate_count=[]
#for col in df.columns:
#    Column.append(col)
#    i=df.duplicated(subset=[col]).sum()
#    Duplicate_count.append(df.duplicated(subset=[col]).sum())
#plt.hist(Duplicate_count,bins=20)

#Identify the number of repetitions of each of the countries
df_duplicate_country=df[df.duplicated()].groupby('Country')['ResponseId'].count()
df_duplicate_country=df_duplicate_country.sort_values(ascending=False)
df_duplicate_country=df_duplicate_country[:10]
n_xaxis = len(df_duplicate_country.index.unique())
fig = plt.figure(figsize=(10, 6))
ax=df_duplicate_country.plot(kind='bar',color='purple')
ax.set_title('Top 10 countries with most number of Duplicates')
ax.set_xticklabels(labels=df_duplicate_country.index, rotation=70, rotation_mode="anchor", ha="right")

#Identify the number of repetitions by employment
df_duplicate_employment=df[df.duplicated].groupby('Employment')['Employment'].count()
df_duplicate_employment=df_duplicate_employment.sort_values(ascending=False)
df_duplicate_employment=df_duplicate_employment[:10]
n_xaxis = len(df_duplicate_employment.index.unique())
fig1 = plt.figure(figsize=(10,6))
ax1=df_duplicate_employment.plot(kind='bar',color='green')
ax1.set_title('Top 10 Employment categories with most number of Duplicates')
ax1.set_xticklabels(labels=df_duplicate_employment.index, rotation=70, rotation_mode="anchor", ha="right")
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here
Columns=df.columns.unique()
df_Duplicate_Removal=pd.DataFrame(columns=['col1','col2'])
for Column in Columns:
    Duplicates=df.duplicated(subset=[Column]).sum()
    new_row=pd.DataFrame({'col1': [Column], 'col2': [Duplicates]})
    df_Duplicate_Removal=pd.concat([df_Duplicate_Removal,new_row],ignore_index=True)

print(df_Duplicate_Removal.sort_values(by='col2',ascending=True).head(45))
```
